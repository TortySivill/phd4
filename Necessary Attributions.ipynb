```python
import gately
import shapley
import numpy as np
import random
import timeit
import shap
```

```python
def f_complex(x):
    # Complex coincedation of the Shapley value and the Gately Point 
    return x[:,0] + 2*(x[:,2] * x[:,1]) - 0.5*(x[:,3] * x[:,4]) + 2*(x[:,4] / x[:,5]) +6*(x[:,6])
```

```python
def f_simple(x):
    # Complex coincedation of the Shapley value and the Gately Point 
    return x[:,0] + 2*(x[:,2]) + 0.8*x[:,1] - 0.5*(x[:,3]) + 0.2*x[:,4] + 2*(x[:,4]) + 3*x[:,5] +6*(x[:,6])
```

```python
def f_sum(x):
    # Complex coincedation of the Shapley value and the Gately Point 
    #return np.sum(np.sign(x),axis=1)
    return np.sign(np.sum(x,axis=1))
```

```python
datas = []
for i in range(10000):
    data = []
    data.append(random.uniform(-0.5,0.5))
    for i in range(0,6):
        data.append(random.uniform(-0.1,0.1))
    datas.append(data)
datas = np.asarray(datas)
y = f_sum(datas)
t = timeit.Timer()       # outside the try/except

starttime = timeit.default_timer()
g_att = gately.Gately(datas[0],datas[1],f_sum)

print("The time difference is :", timeit.default_timer() - starttime)
print(g_att)
print(sum(g_att.values()))

starttime = timeit.default_timer()
att = shapley.shapley_int(f_sum,datas[0],datas[1])
print("The time difference is :", timeit.default_timer() - starttime)
print(att)
print(sum(att.values()))

starttime = timeit.default_timer()
explainer = shap.KernelExplainer(f_sum, datas[1].reshape(1,-1))
shap_values = explainer.shap_values(datas[0].reshape(1,-1), nsamples=100)
print("The time difference is :", timeit.default_timer() - starttime)
print(shap_values)
print(sum(shap_values[0]))
```

```python
selected_inds = 0
ks_wrong = 0
gately_wrong = 0
shap_wrong = 0

ind = 0 

while selected_inds < 200:
    ind += 1
    print(" ")
    
    if f_sum(X_test[ind].reshape(1,-1))[0] != f_sum(X_test[ind+1].reshape(1,-1))[0]:
    

        if f_sum(X_test[ind].reshape(1,-1))[0] > f_sum(X_test[ind+1].reshape(1,-1))[0]:
            sample = X_test[ind]
            ref = X_test[ind+1]
        else:
            sample = X_test[ind+1]
            ref = X_test[ind]

        att = gately.Gately(sample,ref,f_sum)
        
        shap_att = shapley.shapley_int(f_sum,sample,ref)
        
        explainer = shap.KernelExplainer(f_sum, ref.reshape(1,-1))
        shap_values = explainer.shap_values(sample.reshape(1,-1), nsamples=1000)
       
        ks = {}
        for i in range(0,sample.shape[0]):
            ks[i] = shap_values[0][i]

        ks_sorted = sorted(ks, key=ks.get, reverse=True)
        att_sorted = sorted(att, key=att.get, reverse=True)
        shap_sorted = sorted(shap_att, key=shap_att.get, reverse=True)


        if att_sorted != []:
            
            print("gt")
            gts = sample - ref
            gt = {}
            for i in range(0,sample.shape[0]):
                gt[i] = gts[i]
                gt_sorted = sorted(gt, key=gt.get, reverse=True)
            
            gt_indexes = []
            for i in gt_sorted:
                
                x_sample = sample.copy()
                x_sample[i] = ref[i]
                if f_sum(x_sample.reshape(1,-1)) != f_sum(sample.reshape(1,-1)):
                    gt_indexes.append(i)
            print(gt_indexes)
                    
                    
                
            if len(gt_indexes) > 0:
            
        
                indexes = []
                for i in range(0,len(att.values())):
                    if att[i] != 0:
                        indexes.append(list(att.keys())[i])
                print("gately important")
                print(indexes)

                if len(indexes) != len(gt_indexes):
                    gately_wrong += 1

                print("ks important")
                indexes = []
                for i in range(0,len(ks.values())):
                    if ks[i] > 0:
                        indexes.append(list(ks.keys())[i])
                print(indexes)

                if len(indexes) != len(gt_indexes):
                    ks_wrong += 1

                print("shap important")
                indexes = []
                for i in range(0,len(shap_att.values())):
                    if shap_att[i] > 0:
                        indexes.append(list(shap_att.keys())[i])
                print(indexes)

                if len(indexes) != len(gt_indexes):
                    shap_wrong += 1


                selected_inds += 1 

                
```
